```python
import tensorflow as tf
import tensorflow_hub as tfh
import numpy as np
import matplotlib.pyplot as plt
import cv2
```

```python
model = tfh.load('https://tfhub.dev/google/magenta/arbitrary-image-stylization-v1-256/2')
```

```python
def load_image(img_path):
    img = tf.io.read_file(img_path)
    img = tf.image.decode_image(img, channels=3)
    img = tf.image.convert_image_dtype(img, tf.float32)
    img = img[tf.newaxis, :]
    return img
```

```python
style = load_image('greatwave.jpg')
img = load_image('van.jpg')
```

```python
plt.imshow(np.squeeze(style))
plt.show()
```

```python
plt.imshow(np.squeeze(img))
plt.show()
```

```python
modified = model(tf.constant(img), tf.constant(style))[0]
```

```python
plt.imshow(np.squeeze(modified))
plt.show()
```
